```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv('kepler_data.csv', delimiter=',')

# Print column information
print("Columns and data types:")
print(data.dtypes)

# Separate numeric and categorical columns
numeric_cols = data.select_dtypes(include=['number']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

print(f"Numeric columns: {numeric_cols}")
print(f"Categorical columns: {categorical_cols}")

# Check for numeric columns with missing data
missing_numeric_cols = data[numeric_cols].isnull().sum()
print("Columns with missing numeric data:")
print(missing_numeric_cols[missing_numeric_cols > 0])

# Identify columns that will be skipped during imputation
skipped_cols = missing_numeric_cols[missing_numeric_cols == data.shape[0]].index
print(f"Columns to be skipped (no observed values): {skipped_cols}")

# Filter out the columns to be skipped from the list of numeric columns
numeric_cols_filtered = [col for col in numeric_cols if col not in skipped_cols]
print(f"Filtered numeric columns: {numeric_cols_filtered}")

# Handle missing values in numeric columns
numeric_imputer = SimpleImputer(strategy='mean')

# Check for empty DataFrame or issues with columns
print(f"Shape of numeric columns for imputation: {data[numeric_cols_filtered].shape}")

# Handle missing columns and potential empty DataFrame
if data[numeric_cols_filtered].shape[1] == 0:
    raise ValueError("No numeric columns available for imputation.")

# Perform imputation
imputed_numeric_data = numeric_imputer.fit_transform(data[numeric_cols_filtered])

# Check the shape of the imputed data
print(f"Shape of imputed numeric data: {imputed_numeric_data.shape}")

# Convert the imputed data back to a DataFrame
imputed_numeric_df = pd.DataFrame(imputed_numeric_data, columns=numeric_cols_filtered, index=data.index)

# Replace the original numeric columns with imputed values
data[numeric_cols_filtered] = imputed_numeric_df

# Handle missing values in categorical columns
categorical_imputer = SimpleImputer(strategy='most_frequent')
imputed_categorical_data = categorical_imputer.fit_transform(data[categorical_cols])

# Convert the imputed categorical data back to a DataFrame
imputed_categorical_df = pd.DataFrame(imputed_categorical_data, columns=categorical_cols, index=data.index)

# Replace the original categorical columns with imputed values
data[categorical_cols] = imputed_categorical_df

# Check which columns are numeric before scaling
numeric_for_scaling = data.select_dtypes(include=['number']).columns
print(f"Numeric columns for scaling: {numeric_for_scaling}")

# Exclude 'koi_disposition' from feature columns for scaling
feature_columns = numeric_for_scaling.difference(['koi_disposition'])
print(f"Feature columns for scaling: {feature_columns}")

# Handle non-numeric data if any
data_for_scaling = data[feature_columns]

# Convert non-numeric columns to numeric if necessary (this should already be handled)
# Verify the data types
print(data_for_scaling.dtypes)

# Perform scaling
scaled_features = StandardScaler().fit_transform(data_for_scaling)

# Prepare the feature matrix and target vector
X = pd.DataFrame(scaled_features, columns=feature_columns, index=data.index)
y = data['koi_disposition']

# Print shapes to verify
print(f"Shape of DataFrame after imputation: {data.shape}")
print(f"Shape of features (X): {X.shape}")
print(f"Shape of target (y): {y.shape}")
```

```python
# Check for NaNs in the features
print(data[feature_columns].isna().sum())
# Define features and target
X = data[feature_columns]
y = data['koi_disposition']
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize the classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit grid search
grid_search.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import accuracy_score

# Initialize the model (without hyperparameter tuning)
clf = RandomForestClassifier()

# Fit the model
clf.fit(X_train, y_train)

# Predict on the test set
y_pred_before = clf.predict(X_test)

# Calculate accuracy
accuracy_before = accuracy_score(y_test, y_pred_before)
print(f"Accuracy before tuning: {accuracy_before:.4f}")
```

```python
# Get the best estimator from GridSearchCV
best_clf = grid_search.best_estimator_

# Predict on the test set using the best estimator
y_pred_after = best_clf.predict(X_test)

# Calculate accuracy
accuracy_after = accuracy_score(y_test, y_pred_after)
print(f"Accuracy after tuning: {accuracy_after:.4f}")
```

```python
print(f"Accuracy before tuning: {accuracy_before:.4f}")
print(f"Accuracy after tuning: {accuracy_after:.4f}")

# Check if there's an improvement
if accuracy_after > accuracy_before:
    print("The accuracy has improved after hyperparameter tuning.")
else:
    print("The accuracy has not improved after hyperparameter tuning.")
```
