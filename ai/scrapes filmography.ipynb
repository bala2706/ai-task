```python
import requests
import json

def get_actor_filmography(actor_name, api_key):
    try:
        # Search for the actor
        search_url = f"https://api.themoviedb.org/3/search/person?api_key={api_key}&query={actor_name}"
        response = requests.get(search_url)
        response.raise_for_status()  # Raises an HTTPError if the response was an error
        data = response.json()
        
        if data['results']:
            actor_id = data['results'][0]['id']
            # Get actor's movie credits
            credits_url = f"https://api.themoviedb.org/3/person/{actor_id}/movie_credits?api_key={api_key}"
            response = requests.get(credits_url)
            response.raise_for_status()  # Raises an HTTPError if the response was an error
            credits = response.json()
            
            movies = credits.get('cast', [])
            # Sort movies by release date
            sorted_movies = sorted(movies, key=lambda x: x.get('release_date', '1900-01-01'), reverse=True)
            
            # Print the filmography
            print(f"Movies starring {actor_name} in descending order:")
            for movie in sorted_movies:
                title = movie.get('title', 'Unknown')
                release_date = movie.get('release_date', 'Unknown')
                print(f"{title} ({release_date})")
        else:
            print(f"Actor {actor_name} not found.")
    
    except requests.RequestException as e:
        print(f"An error occurred: {e}")

# Get the actor name input from the user
actor_name = input("Enter the name of the actor: ")
api_key = '5ba021eed77e3da7e484c5e7bcd9c8e9'  
get_actor_filmography(actor_name, api_key)
```
